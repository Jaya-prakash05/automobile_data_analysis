# Data Wrangling

```python
#deal with missing data by doing mean on it
avg_normalized_lossess=df['normalized-losses'].astype("float").mean()
avg_normalized_lossess
```

```python
# replace nan with mean
df.loc[:, 'normalized-losses'] = df['normalized-losses'].replace(np.nan, avg_normalized_lossess)
```

```python
avg_bore=df['bore'].astype('float').mean()
df.loc[:,'bore']=df['bore'].replace(np.nan,avg_bore)
df
```

```python
# stroke column
avg_stroke=df['stroke'].astype("float").mean()
df.loc[:,'stroke']=df['stroke'].replace(np.nan,avg_stroke)
df
```

```python
avg_horsepower=df['horsepower'].astype('float').mean()
df.loc[:,'horsepower']=df['horsepower'].replace(np.nan,avg_horsepower)
df
```

```python
avg_peak_rpm=df["peak-rpm"].astype('float').mean()
df.loc[:,'peak-rpm']=df['peak-rpm'].replace(np.nan,avg_peak_rpm)
df
```

```python
avg_num_doors=df['num-of-doors'].value_counts().idxmax()
df.loc[:,"num-of-doors"]=df['num-of-doors'].replace(np.nan,avg_num_doors)
df
```

```python
df=df.dropna(subset=['price'])
df
```

```python
#checking null price values are present or not
is_price_null=df['price'].isnull()
is_price_null.value_counts()
```

```python
#converting data types into proper format
df[['bore','stroke','price','peak-rpm']]=df[['bore','stroke','price','peak-rpm']].astype("float")
df['normalized-losses']=df['normalized-losses'].astype("int")
print(df.dtypes)
```

```python
pd.reset_option('display.max_columns')
```

```python
#data standardization
df['city-L/100km']=235/df['city-mpg']
df["highway-L/100km"]=236/df["highway-mpg"]
df
```

```python
df['length']=df['length']/df['length'].max()
df['width']=df['width']/df['width'].max()
df['height']=df['height']/df['height'].max()
```

```python
df["horsepower"]=df["horsepower"].astype('int')
```

```python
%matplotlib inline
import matplotlib as plt
from matplotlib import pyplot
plt.pyplot.hist(df["horsepower"])

# set x/y labels and plot title
plt.pyplot.xlabel("horsepower")
plt.pyplot.ylabel("count")
plt.pyplot.title("horsepower bins")
```

```python
bins = np.linspace(min(df["horsepower"]), max(df["horsepower"]), 4)
bins
```

```python
group_names = ['Low', 'Medium', 'High']
df['horsepower-binned'] = pd.cut(df['horsepower'], bins, labels=group_names, include_lowest=True )
df[['horsepower','horsepower-binned']].head(20)
```

```python
df["horsepower-binned"].value_counts()
```
